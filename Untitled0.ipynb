```python
import numpy as np
from sklearn.linear_model import LinearRegression

# 훈련 데이터 생성
X = np.array([[1], [2], [3], [4], [5]])  # 입력 변수
y = np.array([2, 4, 6, 8, 10])           # 출력 변수

# 선형 회귀 모델 생성 및 학습
model = LinearRegression()
model.fit(X, y)

# 새로운 입력 데이터에 대한 출력 값 예측
new_X = np.array([[6], [7]])
predicted_y = model.predict(new_X)

print(predicted_y)
```
