### 3. Local 에서 직접 data 를 불러오도록 하자

```python
iris = pd.read_csv("/content/drive/MyDrive/ML/실습 2주차/iris.csv")
iris.head()
```

#### Data PreProcessing

##### 1. Split X and Y

```python
X_iris = iris[["sepal.length","sepal.width","petal.length","petal.width"]]
y_iris = iris[["variety"]]
```

Column 명을 변경

```python
# 컬럼명 변경 -> 'sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'
X_iris_columns = ['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)'] #list
X_iris.columns = X_iris_columns
X_iris.head(5)
```

```python
y_iris
```

```python
y_iris['variety'].unique() #unique() : 중복되는 값없이 출력
```

### 3. Split train set, test set

```python
# 함수를 이용해서 나누기
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_iris,y_iris,test_size=0.3,random_state=0)
#train할 data,test할 data, traing할 정답,test할 정답
```

### 4. Machine learning modeling

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3) #n_neighbors = k
```

### 5. Training

```python
X_train #randomly output
```

```python
y_train
```

```python
## knn 알고리즘 traiging
knn.fit(X_train,y_train)
```

```python
print(y_train.shape)
print(y_train.values.ravel().shape)
#(105,1)matrix와 (105,)matrix의 차이
```

```python
## knn 알고리즘 traiging
knn.fit(X_train,y_train.values.ravel())
```

### 6. Predict

```python
import numpy as np

# 'sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)' 을 임의로 정한다
# 5.3, 2.1, 1.3, 0.2 의 수치로 정함
Our_X = np.array([[5.3,2.1,1.3,0.2]]) #다수의 데이터입력을 위해 2차배열로 생성
Our_X
```

```python
# 임의의 데이터로 prediction
prediction = knn.predict(Our_X)
print("예측 : {}".format(prediction))
```

### 7. Accuracy

```python
y_pred = knn.predict(X_test)
print(y_pred.shape)
print("테스트 세트에 대한 예측값:\n {}".format(y_pred))
```

```python
# y_test 를 1차원 matrix 로 변경
y_test_flatten = np.array(y_test).flatten()
```

```python
y_test.shape
```

```python
y_test_flatten.shape
```

```python
y_pred == y_test_flatten
```

```python
print("테스트 세트의 정확도: {:.2f}".format(np.mean(y_pred == y_test_flatten))) #true와 false의 mean으로 정확도 측정
```

```python
print("테스트 세트의 정확도: {:.2f}".format(knn.score(X_test, y_test_flatten)))
```
